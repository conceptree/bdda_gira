```python
# Read the datasets

Gira_H1 = pd.read_csv('Gira_H1.csv')
Gira_H2 = pd.read_csv('Gira_H2.csv')
Meteo = pd.read_csv('Lx_meteo.csv')
```

```python
# Concatenate Gira Tables
Concatenate = [Gira_H1, Gira_H2]

gira = pd.concat(Concatenate)
```

```python
gira.head()
```

```python
gira.info()
```

```python
gira.dropna()
```

```python
gira['ID'] = gira['desigcomercial'].apply(lambda x: x.split('-')[0])
```

```python
gira.head()
```

```python
gira['station'] = gira['desigcomercial'].apply(lambda x: x.split('-')[1])
gira = gira.drop('desigcomercial', axis=1)
```

```python
gira.head()
```

```python
gira.rename(columns = {'numbicicletas':'bikes_avail', 'numdocas':'total_docks', 'entity_ts':'timestamp', 'estado':'state'}, inplace = True)
```

```python
gira.head()
```

```python
cols = ['ID','timestamp','station','bikes_avail','total_docks','state','position']
```

```python
gira = gira[cols]
```

```python
gira.head(20)
```

```python
Meteo.info()
```

```python
Meteo = Meteo.drop(columns = ['name','tempmax','tempmin','feelslikemax','feelslikemin','dew','humidity','precipprob',
                 'precipcover','preciptype','snow','snowdepth','windgust','windspeed','winddir','sealevelpressure','cloudcover',
                 'visibility','solarradiation','solarenergy','solarenergy','uvindex','severerisk','sunrise','sunset','moonphase',
                 'conditions','description','icon','stations'], axis = 1)
```

```python
Meteo.head()
```

```python
Meteo.rename(columns = {'datetime':'date'}, inplace = True)
```

```python
Meteo.head()
```

```python
gira['date'] = gira['timestamp'].apply(lambda x: x.split('T')[0])
```

```python
gira.head()
```

```python
gira['time'] = gira['timestamp'].apply(lambda x: x.split('T')[1])
```

```python
gira.head()
```

```python
df = pd.merge(gira, Meteo, on = 'date')
```
